```python
# pip install transformers accelerate

from transformers import AutoTokenizer
import transformers
import torch

model = "varshil27/Llama-2-7b-finetune-1"
prompt = "These are patient's Symptoms Cough, Throat pain , Headache. What are the top 5 most likely diseases corresponding to these symptoms? Output only disease names."

tokenizer = AutoTokenizer.from_pretrained(model)
pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    torch_dtype=torch.float16,
    device_map="auto",
)

sequences = pipeline(
    f'<s>[INST] {prompt} [/INST]',
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
    max_length=200,
)
for seq in sequences:
    print(f"Result: {seq['generated_text']}")
```
